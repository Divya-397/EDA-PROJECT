## Cleaning Data 

### Removing Unwanted columns

```python
df = df.drop([  'Classification Key', 'Historic', 'External IDs','Admin Units', 'CPI','Event Name','ISO','Latitude','Longitude','Magnitude','Magnitude Scale'], axis=1)
```

```python
df = df.drop(columns=['AID Contribution (\'000 US$)'])
df = df.drop(columns=['Reconstruction Costs (\'000 US$)'])
df = df.drop(columns=['Reconstruction Costs, Adjusted (\'000 US$)'])
df = df.drop(columns=['DisNo.'])
```

### Filling the null values

```python
df.fillna(0, inplace=True)
```

```python
df.info()
```

### Renaming of column headings ###

```python
df.rename(columns={'Insured Damage (\'000 US$)':'Insured_Damage','Insured Damage, Adjusted (\'000 US$)':'Insured_Damage Adjusted','Total Damage (\'000 US$)': 'Total Damage','Total Damage, Adjusted (\'000 US$)':'Total Damage Adjusted'})
```

### Removing spaces in column headings

```python
df.columns = [col.replace(' ', '_') for col in df.columns]
```

```python
df.isnull().sum()
```

### Finding dupicates

```python
df.duplicated()
df.drop_duplicates(inplace=True)
```

### Correction of date format

```python
df['Entry_Date'] = pd.to_datetime(df['Entry_Date'], dayfirst=True).dt.strftime('%Y-%m-%d')
df['Last_Update'] = pd.to_datetime(df['Last_Update'], dayfirst=True).dt.strftime('%Y-%m-%d')
```

### unique values ###

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in column '{column}': {len(unique_values)}")
    print(unique_values)
    print()
unique_counts = df.nunique()
print(unique_counts)

```

## Statistical analysis

```python
df.describe()
```

###  Correlation ###

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(9, 9))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
covariance = df['Total_Deaths'].cov(df['Total_Damage_(\'000_US$)'])
print('Covariance:', covariance)
correlation = df['Total_Deaths'].corr(df['Total_Damage_(\'000_US$)'])
print('Correlation:', correlation)
```

### 1.Frequency count of categorical columns

```python
columns_to_visualize = ['Disaster_Group', 'Disaster_Subgroup', 'Disaster_Type','Subregion','Region']
for column in columns_to_visualize:
    plt.figure(figsize=(10, 6))
    value_counts = df[column].value_counts()
    sns.barplot(x=value_counts.values, y=value_counts.index, palette="viridis")
    
    plt.title(f"Value Counts for {column}")
    plt.xlabel("Counts")
    plt.ylabel(column)
    plt.show()
```

### 2.Disaster group with highest number of total death ###

```python
death_by_group = df.groupby('Disaster_Group')['Total_Deaths'].sum()
death_by_group_sorted = death_by_group.sort_values(ascending=False)
plt.figure(figsize=(5,5))
sns.barplot(x=death_by_group_sorted.index, y=death_by_group_sorted.values, palette='viridis')
plt.title('Disaster Groups with the Highest Number of Total Deaths', fontsize=16)
plt.xlabel('Disaster Group', fontsize=10)
plt.ylabel('Total Deaths', fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 3.Distribution of disaster types


```python
disaster_types = df['Disaster_Type'].value_counts()
disaster_percentage = (disaster_types / disaster_types.sum()) * 100
filtered_disaster_types = disaster_percentage[disaster_percentage > 4]
plt.figure(figsize=(5, 5))
plt.pie(
    filtered_disaster_types.values, 
    labels=filtered_disaster_types.index, 
    autopct='%1.1f%%', 
    startangle=90
)
plt.title('Distribution of Disaster Types (>4%)')
plt.show()
print("Filtered Distribution of Disaster Types (>4%):")
print(filtered_disaster_types)
```

### 4.Region with most Disaster occured

```python
disasters_by_subregion = df['Region'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=disasters_by_subregion.values[:10], y=disasters_by_subregion.index[:10], palette="crest")
plt.title("Top 5 Regions with Most Disasters")
plt.xlabel("Number of Disasters")
plt.ylabel("Region")
plt.show()
```

### 5. How do disasters vary across subregions


```python
disasters_by_subregion = df['Subregion'].value_counts()
plt.figure(figsize=(12, 8))
sns.barplot(x=disasters_by_subregion.values[:10], y=disasters_by_subregion.index[:10], palette="crest")
plt.title("Top 10 Subregions with Most Disasters")
plt.xlabel("Number of Disasters")
plt.ylabel("Subregion")
plt.show()
```

### 6.Most Frequent Disaster Types in Top Subregion

```python
top_subregion = ['Sub-Saharan Africa', 'Latin America and the Caribbean', 'Southern Asia']
df_filtered = df[df['Subregion'].isin(top_subregion)]
disaster_counts = df_filtered.groupby(['Subregion', 'Disaster_Type']).size().reset_index(name='Count')
top_disasters = disaster_counts.loc[disaster_counts.groupby('Subregion')['Count'].idxmax()]
plt.figure(figsize=(6, 6))
sns.barplot(data=top_disasters, x='Subregion', y='Count', hue='Disaster_Type', dodge=False)
plt.title('Most Frequent Disaster Types in Top Subregion')
plt.xlabel('Subregion')
plt.ylabel('Disaster Count')
plt.legend(title='Disaster Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 7.which countries are most prone to diaster

```python
country_counts = df['Country'].value_counts().head(20)
plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette="plasma")
plt.title("Top 20 Countries with Most Disasters")
plt.xlabel("Number of Disasters")
plt.ylabel("Country")
plt.show()
```

### 8.Most Frequent Disaster Types in Top Countries

```python
top_countries = ['China', 'India', 'United States of America']
df_filtered = df[df['Country'].isin(top_countries)]
disaster_counts = df_filtered.groupby(['Country', 'Disaster_Type']).size().reset_index(name='Count')
top_disasters = disaster_counts.loc[disaster_counts.groupby('Country')['Count'].idxmax()]
plt.figure(figsize=(6, 6))
sns.barplot(data=top_disasters, x='Country', y='Count', hue='Disaster_Type', dodge=False)
plt.title('Most Frequent Disaster Types in Top Countries')
plt.xlabel('Country')
plt.ylabel('Disaster Count')
plt.legend(title='Disaster Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 11.Common Disaster Type and their Subtype Counts

```python
disaster_types = ['Flood', 'Road', 'Storm']
result = {}

for disaster_type in disaster_types:
    filtered_data = df[df['Disaster_Type'] == disaster_type]
    subtype_counts = filtered_data['Disaster_Subtype'].value_counts()
    result[disaster_type] = subtype_counts

for disaster_type, subtype_counts in result.items():
    print(f"\nCounts of Disaster Subtypes for Disaster Type: {disaster_type}")
    print(subtype_counts)
```

### 12.Total Damages Across Disaster Subgroups

```python
damage_by_subgroup = df.groupby('Disaster_Subgroup')[['Insured_Damage_(\'000_US$)', 
                                                      'Insured_Damage,_Adjusted_(\'000_US$)', 
                                                      'Total_Damage_(\'000_US$)', 
                                                      'Total_Damage,_Adjusted_(\'000_US$)']].sum()

plt.figure(figsize=(8,8))
sns.heatmap(damage_by_subgroup, annot=True, fmt='.2f', cmap='viridis', cbar_kws={'label': 'Damage in $000s'})
plt.title('Heatmap of Total Damages Across Disaster Subgroups', fontsize=16)
plt.xlabel('Damage Type', fontsize=12)
plt.ylabel('Disaster Subgroup', fontsize=12)
plt.tight_layout()
plt.show()
```

### 13.Which river basins experience the highest number of disasters

###  18. Most common Disaster type and their Associated types counts


```python
target_disasters = ['Flood', 'Storm', 'Earthquake','Drought']
filtered_df = df[df['Disaster_Type'].isin(target_disasters)]
fig, axes = plt.subplots(1, 4, figsize=(18, 6), sharey=True)
colors = ['skyblue', 'orange', 'green','red']
for idx, disaster in enumerate(target_disasters):
    disaster_df = filtered_df[filtered_df['Disaster_Type'] == disaster]
    associated_counts = disaster_df['Associated_Types'].value_counts().head(10)
    sns.barplot(x=associated_counts.values, y=associated_counts.index, ax=axes[idx], color=colors[idx])
    axes[idx].set_title(f"{disaster} - Associated Types", fontsize=14)
    axes[idx].set_xlabel('Count', fontsize=12)
    axes[idx].set_ylabel('Associated Types', fontsize=12)
plt.tight_layout()
plt.show()
```
